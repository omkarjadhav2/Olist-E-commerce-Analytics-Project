```python
#import Pandas
import pandas as pd
```

```python
# Read the dataframe
item_df = pd.read_csv("olist_order_items_dataset.csv")
```

```python
# Check structure
item_df.head()
```

```python
# Check data types and null count
item_df.info()
```

```python
# check unique values
item_df.nunique()
```

```python
# summary statistics
item_df.describe()
```

```python
# check null values
item_df.isnull().sum()
```

```python
# check duplicate values
item_df.duplicated().sum()
```

```python
# column drop
item_df.drop(["shipping_limit_date"], axis=1,inplace=True)
```

```python
# product dataframe
all_products = pd.read_csv("olist_products_dataset.csv")
```

```python
# merge the dataframe
item_df = item_df.merge(
    all_products,
    on="product_id",
    how="left"
)
```

```python
item_df.info()
```

```python
# drop unnecessary columns
item_df.drop(columns=["product_name_lenght","product_description_lenght",
                      "product_photos_qty","product_weight_g","product_length_cm",
                      "product_height_cm","product_width_cm"

], axis=1,inplace=True)
```

```python
# translation dataframe
translation_df = pd.read_csv("product_category_name_translation.csv")
```

```python
# merge translation dataframe
item_df = item_df.merge(
    translation_df,
    on="product_category_name",
    how="left"
)
```

```python
# Rename column
item_df["product_category_name"] = item_df['product_category_name_english']
```

```python
item_df.drop(["product_category_name_english"], axis=1,inplace=True)
```

```python
# Remove whitespaces
item_df["order_id"] = item_df["order_id"].str.strip()
item_df["product_id"] = item_df["product_id"].str.strip()
item_df["seller_id"] = item_df["seller_id"].str.strip()
```

```python
# converts all values to lowercase
item_df["order_id"] = item_df["order_id"].str.lower()
item_df["product_id"] = item_df["product_id"].str.lower()
item_df["seller_id"] = item_df["seller_id"].str.lower()
```

```python
item_df.head()
```

```python
# export data to csv
item_df.to_csv("cleaned_order_items.csv", index=False)
```
